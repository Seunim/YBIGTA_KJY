```python
import pandas as pd
import selenium
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from html_table_parser import parser_functions


url='https://www.yonsei.ac.kr/sc/support/calendar.jsp'

request=requests.get(url)
soup=BeautifulSoup(request.text,"html.parser")
data=soup.find("table",{"class":"tblH ty2 jw_vertical"})
table=parser_functions.make2d(data)

df= pd.DataFrame(data=table[1:],columns=table[0])
df
 

```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from PIL import Image
import urllib.request

# 해당 URL로 이동
driver= webdriver.Chrome()
driver.get('https://naver.com') 

# 웹 페이지 요소 로드를 기다림
driver.implicitly_wait(5)
search=driver.find_element(By.XPATH, '//*[@id="query"]')
```

```python
# 검색어 입력
search.send_keys('고려대학교')
search.send_keys(Keys.RETURN)
```

```python
img = driver.find_element(By.XPATH, '//*[@id="main_pack"]/div[2]/div[2]/div[1]/div/div[2]/a/img')
img_url = img.get_attribute('src')
```

```python
url_openimg=Image.open(urllib.request.urlopen(img_url))
url_openimg
```

```python
# 브라우저 닫기
driver.close()
```
