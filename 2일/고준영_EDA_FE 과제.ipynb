### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('/content/drive/MyDrive/airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12
df.drop(columns = df.columns[df.isnull().sum()/len(df) >= 0.25], axis = 1, inplace = True)
df.dropna(subset = df.columns[df.isnull().sum()/len(df) >= 0.1], inplace = True)
df.fillna(df.median(numeric_only=True), inplace = True)
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1)
Identified=[]
for _ in df['host_identity_verified']:
  for j in df['host_has_profile_pic']:
    for k in df['host_verifications']:
      if (_ =='t') and (j =='t') and (k=='t'):
        Identified.append('T')
      else:
        Identified.append('F')
df['Host_identification'] = Identified

# 숙소 예약시 거래 안전 등의 요소 파악을 위한 호스트에 대한 보수적인 평가 기준
```

```python
#Q13-2)
QuickBuisness=[]
for _ in df['room_type']:
  for j in df['instant_bookable']:
    if (_=='Hotel room') and (j == 't'):
      QuickBuisness.append('T')
    else:
      QuickBuisness.append('F')
df['QuickBuisness']= QuickBuisness

# 빠른 예약이 필요한 업무용 호텔 리스트 파악용
```

```python
#Q13-3)
Anonymous=[]
for _ in df['requires_license']:
  for j in df['require_guest_profile_picture']:
    for k in df['require_guest_phone_verification']:
      if (_ == 'f') and (j == 'f') and (k == 'f'):
        Anonymous.append('T')
      else:
        Anonymous.append('F')
df['Anonymous']=Anonymous
#익명으로 묵을 수 있는 숙소, 불법적 용도 관련 리스트업
```
