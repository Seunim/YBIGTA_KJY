```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# data set: kaggle <Titanic - Machine Learning from Disaster>
# https://www.kaggle.com/competitions/titanic/overview
df = pd.read_csv('/content/drive/MyDrive/train.csv')
df
```

```python
fig = plt.subplots(1, 1, figsize=(10, 10))

df_scatter = df.copy()
df_scatter = df_scatter[df_scatter['Sex'] == 'male']

plt.rc('font', size=15)

# Bubble Chart
sns.scatterplot(data = df_scatter, x='Age', y='Fare', size='Parch', hue='Survived', )

plt.show()
```
